```python
import pandas as pd
import random as rd
df = pd.read_csv("centroids_ksize_binary.csv")

colnames = df.columns
gd1 = ['argc','safe_math','packed_struct','unions','volatile_pointers']
gd2 = ['embedded_assigns','pre_incr_operator','pre_decr_operator','unary_plus_operator','int8','uint8','float','math64']
```

```python
dflist = []
for idx,row in df.iterrows():
    lrow = list(row)
    rlist = []
    for i,v in enumerate(lrow):
        r = rd.uniform(0,1)
        if (v == 0):
            if colnames[i] in gd1:
                r = rd.uniform(0.0,0.25)
            elif colnames[i] in gd2:
                r = rd.uniform(0.001,0.01)
            else:
                r = rd.uniform(0.0001,0.001)
        elif (v==1):
            if colnames[i] in gd1:
                r = rd.uniform(0.75,1.0)
            elif colnames[i] in gd2:
                r = rd.uniform(0.65,0.75)
            else:
                r = rd.uniform(0.50,1.0)
        rlist.append(r)
    dflist.append(rlist)
    
print(dflist)
```

```python
data = pd.DataFrame(dflist, columns=colnames)
data.to_csv("centroids_ksize_prob.csv",index=False)
```
