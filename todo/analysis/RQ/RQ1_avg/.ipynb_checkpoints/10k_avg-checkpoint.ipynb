```python
import pandas as pd
fpath="kconfig_10k_all.csv"
df = pd.read_csv(fpath)
df = df.drop('RN', axis=1, inplace=False)
df = df[df.ST != "ksize"]
df
```

```python
df2 = df
df2 = df2.groupby(['VR', 'ST']).sum().reset_index()
df2
```

```python
df3 = df2
cols = df3.columns [2:]
df3[cols] =  round (df3[cols] / 3)
df3[cols] = df3[cols].astype(int)
```

```python
i=0
for index, row in df3.iterrows():
    i=i+1
    print(" & ".join(map(str,list(row))),"\\\\\n\\hline")
    if i%3 == 0: print("\n\hline\n")
```
