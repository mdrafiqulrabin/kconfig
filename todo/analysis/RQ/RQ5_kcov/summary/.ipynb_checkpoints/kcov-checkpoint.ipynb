```python
import pandas as pd

cv = ['statement','branch']
for c in cv:
    df = pd.read_csv(c+".csv")
    df['cov'] = (df['per'] * df['num']) / 100

    ts = ["regression","default","swarm","kconfig"]

    print("\n"+c+" coverage:")
    for t in ts:
        dfd = df.loc[df['ts'] == t]
        r = dfd['cov'].sum()/dfd['num'].sum()
        print(t,"= ",r,"[",int(dfd['cov'].sum()),"/",int(dfd['num'].sum()),"]")
        
        
df = pd.read_csv("function.csv")
df['cov'] = (df['per'] * df['num']) / 100
df.loc[df['cov'] > 0, 'cov'] = 1

ts = ["regression","default","swarm","kconfig"]

print("function coverage:")
for t in ts:
    dfd = df.loc[df['ts'] == t]
    r = dfd['cov'].sum()/len(dfd.index)
    print(t,"= ",r)
```
