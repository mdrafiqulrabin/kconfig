```python
import pandas as pd
import glob

tslist = ['statement','branch','function']
for tsel in tslist:
    ftxt = glob.glob("*"+tsel+".txt")
    lcsv = []
    for fn in ftxt: 
        sfn = (fn.split(".")[0]).split("_")
        lines = []
        with open(fn) as f:
            lines = [line.rstrip('\n') for line in f]
        item = ""
        for w in lines:
            item = ""
            w = w.split(",")
            item = [sfn[1],sfn[0],w[0],w[1]]
            lcsv.append(item)
    df = pd.DataFrame(lcsv)
    df.columns = ['ts','vn','per','num']
    df.to_csv(tsel+".csv",index=False)
```

```python
import pandas as pd
import numpy as np

ts = ["regression","default","swarm","kconfig"]
vr = ['482','540','610','710']
cv = ['statement','branch','function']

for v in vr:
    for t in ts:
        sbf=[]
        for c in cv:
            df = pd.read_csv(c + ".csv")
            df['cov'] = (df['per']*df['num'])/100
            df = df[(df['ts']==t) & (df['vn']==int(v))]
            if c == cv[2]:
                df.loc[df['cov'] > 0, 'cov'] = 1
            else:
                df['cov'] = df['cov'].apply(lambda x: round(x))
            sbf.append(int(df['cov'].sum()))
        label = str(v) + " & " + str(t) + " & " + str(sbf[0]) + " & " + str(sbf[1]) + " & " + str(sbf[2]) + " \\\\"
        print(label)
        print("\hline")
    print("\hline")
```
